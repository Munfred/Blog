```python
features = ln.Feature.lookup()
```

```python
for i_, r in tqdm(file_annotations.iterrows()):
    artifact = ln.Artifact.filter(uid = r['uid']).one()
    
    for o_ in r['Organism'].split(', '):
        organism = ln.ULabel.filter(name = o_).one()
        artifact.labels.add(organism, feature = features.organism)
    
    for t_ in r['Tissue'].split(', '):
        tissue = ln.ULabel.filter(name = t_).one()
        artifact.labels.add(tissue, feature = features.tissue)
    
    for t_ in r['Technique'].split(' & '):
        technique = ln.ULabel.filter(name = t_).one()
        artifact.labels.add(technique, feature = features.technique)
    
    artifact.save()
```

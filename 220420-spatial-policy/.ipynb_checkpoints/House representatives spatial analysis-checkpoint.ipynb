```python
from glob import glob
from xml.etree import ElementTree as ET
```

```python
call_descrs = {}
vote_result = {}

for fname in glob('xmls/roll*.xml'):
    with open(fname, 'rb') as fh:
        text = fh.read()
        root = ET.XML(text)

        meta, votes = root

        for m in meta:
            if m.tag == 'rollcall-num':
                rollcall_num = m.text

            if m.tag == 'vote-desc':
                vote_desc = m.text
                
        call_descrs[rollcall_num] = vote_desc

        name_votes = {}
        for vote in votes:
            leg, vot = vote

            name_id = leg.get('name-id')
            vote_text = vot.text
            
            name_votes[name_id] = vote_text
            
        vote_result[rollcall_num] = name_votes
```

```python
v_res = pd.DataFrame(vote_result).T
v_res.iloc[:20, :10]
```

```python
c_descr = pd.Series(call_descrs).to_frame('description')
c_descr.head(10)
```
